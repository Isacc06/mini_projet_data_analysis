# Mini projet data proc 

```python
import pandas as pd
import requests as req  
import zipfile
import io   
import os
```

```python
# 1. Créer la structure de répertoires
base_dir = "C:/Users/hp/Desktop/mini_projet_data_analysis"
sub_dirs = ["data", "output", "code"]

# Créer le répertoire principal et les sous-répertoires
os.makedirs(base_dir, exist_ok=True)
for sub_dir in sub_dirs:
    os.makedirs(os.path.join(base_dir, sub_dir), exist_ok=True)

print(f"Structure de répertoires créée dans '{base_dir}'")
```

```python
# -*- coding: utf-8 -*-

url = "https://download.geonames.org/export/dump/countryInfo.txt"

# Lire en spécifiant le caractère de commentaire
df = pd.read_csv(url,sep='\t', header=None,encoding='utf-8',comment='#',on_bad_lines='skip'
             ,names=['ISO', 'ISO3', 'ISO_Numeric', 'fips', 'country', 'capital', 'area_sq_km',
                     'population', 'continent', 'tld', 'currency_code', 'currency_name',
                     'phone', 'postal_code_format', 'postal_code_regex', 'languages',
                     'geoname_id', 'neighbours', 'equivalent_fips_code'])

df.head()

# afficher le code du Burkina Faso
ISO = df[df['country'] == 'Burkina Faso']['ISO'].values[0]
```

```python
#Télécharger le fichier zip avec code BF

url = "https://download.geonames.org/export/dump/"+ISO+".zip"
output_path = os.path.join(base_dir, "data", "BF.zip")
    
    # Télécharger le fichier
response = req.get(url)
with open(output_path, 'wb') as f:
        f.write(response.content)
    
    # Extraire le fichier
with zipfile.ZipFile(output_path, 'r') as zip_ref:
    zip_ref.extractall(os.path.join(base_dir, "data"))

    print(f"Fichier téléchargé et extrait dans '{os.path.join(base_dir, 'data')}'")
            
```

```python

# lister les fichiers dezipé dans le répertoire data
for file in os.listdir(os.path.join(base_dir, "data")):
    print(file)
```

```python
BF_data = pd.read_csv(os.path.join(base_dir, "data", "BF.txt"), sep='\t', header=None, encoding='utf-8', comment='#')

BF_data.head()
```

```python
# select the columns of interest
cols = [0, 1,  4, 5 ]

burkina_location = BF_data[cols]

burkina_location.columns = ['ID', 'location_name', 'lat', 'long']

burkina_location.head()
```

```python
# sauvegarder le fichier dans le répertoire output

output_file = os.path.join(base_dir, "output", "burkina_location.csv")
burkina_location.to_csv(output_file, index=False)
```

```python
# les données d'une localité précise
nom_localite = 'gounghin'

localite_data = burkina_location[burkina_location['location_name'].str.lower() == nom_localite.lower()]
localite_data

# sauvegarder les données de la localité dans un fichier CSV
localite_output_file = os.path.join(base_dir, "output", f"{nom_localite}.csv")
```

```python
# selection des localités avec les première lettres commencant entre A ET P par orde alphabétique

localites_A_P = burkina_location[burkina_location['location_name'].str[0].str.upper().between('A', 'P')]
localites_A_P_sorted = localites_A_P.sort_values(by='location_name')
localites_A_P_sorted
```

```python
# identifiez la latitude, la longitude minimale et les noms de lieux correspondants
min_lat = localites_A_P_sorted['lat'].min()
min_long = localites_A_P_sorted['long'].min()

print(f"Latitude minimale: {min_lat}, Longitude minimale: {min_long}")
```

```python
# Filtrer les localités avec latitude => 11 et longitude <= 0.5

filtered_localites = localites_A_P_sorted[(localites_A_P_sorted['lat'] >= 11) & (localites_A_P_sorted['long'] <= 0.5)]
filtered_localites
```

```python
# Enregistrement dans une feuille excel dans le 

with pd.ExcelWriter('output/mini_projet.xlsx',) as writer:
    localite_data.to_excel(writer, sheet_name= nom_localite, index=False)
    localites_A_P_sorted.to_excel(writer, sheet_name='A_to_P', index=False)
```
